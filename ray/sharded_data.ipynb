```python
import ray
# ray.shutdown()
ray.init(ignore_reinit_error=True)
ray.available_resources()
```

Create a data sample with the forumula y = x**2 with file names data_{partition} in parquet format.


```python
import pandas as pd
import numpy as np

# Parameters
num_partitions = 2
rows_per_partition = 1000

for partition in range(num_partitions):
    x = np.linspace(partition * 10, (partition + 1) * 10, rows_per_partition)
    y = x ** 2 + 10
    df = pd.DataFrame({'x': x, 'y': y})
    df.to_parquet(f"data_{partition}.parquet")
```

```python
import xgboost
from ray.train import ScalingConfig
from ray.train.xgboost import XGBoostTrainer
import pandas as pd
import ray.data

# Load all partitions as a Ray Dataset
data = ray.data.read_parquet([f"data_{i}.parquet" for i in range(2)])

params = {
    "tree_method": "approx",
    "objective": "reg:squarederror",
    "eta": 0.05,
    "subsample": 0.5,
    "max_depth": 4,
}

scaling_config = ScalingConfig(num_workers=2, resources_per_worker={"CPU": 4})
trainer = XGBoostTrainer(
    label_column="y",
    params=params,
    datasets={"train": data},
    scaling_config=scaling_config,
    num_boost_round=500,
 )

result = trainer.fit()
```
